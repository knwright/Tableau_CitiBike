```python
# Dependencies
import glob
import pandas as pd
```

```python
# Use glob and pandas to create two dataframes, one for each set of data,
# membership and tripdata
memberData_clean = pd.concat([pd.read_csv(f) for f in glob.glob('data/2018_memberdata/*.csv')], 
                             ignore_index = True)

tripData_clean = pd.concat([pd.read_csv(f) for f in glob.glob('data/2018_tripdata/*.csv')],
                          ignore_index = True)
```

```python
# View memberData_clean df
memberData_clean.head()
```

```python
# Count member data entries per column
memberData_clean.count()
```

```python
# View tripData_clean df
tripData_clean.head()
```

```python
# Count trip data entries per column
tripData_clean.count()
```

```python
# Export cleaned data to csv
memberData_clean.to_csv("data/2018_memberdata/2018_memberData_clean.csv")

tripData_clean.to_csv("data/2018_tripdata/2018_tripData_clean.csv")
```
